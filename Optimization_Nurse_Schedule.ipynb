```python
from gurobipy import *
import pandas as pd
import numpy as np
```

```python
m=Model()
```

```python
#staff i 是否上白班
Di= m.addVars(22,vtype=GRB.BINARY)
#staff i 是否上中班
Gi= m.addVars(22,vtype=GRB.BINARY)
#staff i 是否上早晚班
Li= m.addVars(22,vtype=GRB.BINARY)
#staff i 是否上大夜班
Ni= m.addVars(22,vtype=GRB.BINARY)
```

```python
#staff i 是否上day j 的白班
dij= m.addVars(22,30,vtype=GRB.BINARY)
#staff i 是否上day j 的中班
gij= m.addVars(22,30,vtype=GRB.BINARY)
#staff i 是否上day j 的早晚班
lij= m.addVars(22,30,vtype=GRB.BINARY)
#staff i 是否上day j 的夜班
nij= m.addVars(22,30,vtype=GRB.BINARY)
```

```python
#if 中班staff i works on day j 的大夜班
ngij=m.addVars(22,30,vtype=GRB.BINARY)
#if 大夜班staff i works on day j 的中班
gnij=m.addVars(22,30,vtype=GRB.BINARY)
#if 早晚班staff i works on day j 的中班
glij=m.addVars(22,30,vtype=GRB.BINARY)
#if 早班staff i works on day j 的中班
gdij=m.addVars(22,30,vtype=GRB.BINARY)
#if 早班staff i works on day j 的大夜班
ndij=m.addVars(22,30,vtype=GRB.BINARY)
#if 夜班staff i works on day j 的早班
dnij=m.addVars(22,30,vtype=GRB.BINARY)
#if 早晚班staff i works on day j 的早班
dlij=m.addVars(22,30,vtype=GRB.BINARY)
#if 中班staff i works on day j 的早班
dgij=m.addVars(22,30,vtype=GRB.BINARY)
#Total working hours of staff i
Ui=m.addVars(22,vtype=GRB.INTEGER)
#Total off-day working hours of staff i on day j
tij=m.addVars(22,30,vtype=GRB.INTEGER)
```

```python
Ci=[]
```

```python
Si = [90000,90000,87000,87000,85000,83000,78000,78000,75000,72000,70000,68000,68000,68000,65000,65000,65000,65000,63000,62000,58000,58000]
```

```python
for i in range(22):
    Ci.append(round(Si[i]/144,0))
```

```python
Ci
```

```python
m.setObjective( quicksum([Si[i]*(Di[i]+Gi[i]+Li[i]+Ni[i])for i in range(22)]) + quicksum([0.33*Ci[i]*(Ui[i]-18*8) for i in range(22)]) 
               + quicksum([0.33*Ci[i]*quicksum([tij[i,j] for j in range(30)]) for i in range(22)]) + quicksum([0.53*Ci[i]*quicksum([ndij[i,j]+ngij[i,j]+nij[i,j] for j in range(30)]) for i in range(22)]))
```
